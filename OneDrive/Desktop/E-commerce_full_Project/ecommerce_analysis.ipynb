```python
query_final = """
SELECT 
    u.user_id,
    u.name,
    u.gender,
    u.city,
    u.signup_date,
    o.order_id,
    o.order_date,
    o.order_status,
    p.product_name,
    p.category,
    oi.quantity,
    oi.item_price,
    oi.item_total
FROM users u
JOIN orders o ON u.user_id = o.user_id
JOIN order_items oi ON o.order_id = oi.order_id
JOIN products p ON oi.product_id = p.product_id
"""
df = pd.read_sql(query_final, conn)

df.head()
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df["signup_date"] = pd.to_datetime(df["signup_date"])
```

```python
df["month"] = df["order_date"].dt.month_name()
df["year"] = df["order_date"].dt.year
```

```python
monthly_revenue = df.groupby("month")["item_total"].sum()
monthly_revenue
```

```python
df.head()
```

```python
top_products = df.groupby("product_name")["item_total"].sum().sort_values(ascending=False).head(10)
top_products
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="darkgrid")

plt.rcParams["grid.color"] = "lightgrey"
plt.rcParams["grid.linestyle"] = "--"
plt.rcParams["grid.alpha"] = 0.6
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=monthly_revenue,
    x="order_date",
    y="item_total",
    palette="viridis"
)

plt.title("Monthly Revenue Trend", fontsize=16, fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.ticklabel_format(style='plain', axis='y') 

plt.xticks(rotation=45)

plt.tight_layout()

plt.savefig("monthly_revenue.png", dpi=300)

plt.show()
```

```python
top_products = (
    df.groupby("product_name")["item_total"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

plt.figure(figsize=(10,6))

sns.barplot(
    data=top_products,
    y="product_name",
    x="item_total",
    palette="mako"
)

plt.title("Top 10 Products by Revenue", fontsize=16, fontweight="bold")
plt.xlabel("Revenue")
plt.ylabel("Product")

plt.tight_layout()

plt.savefig("top_products.png", dpi=300)

plt.show()
```

```python
category_sales = (
    df.groupby("category")["item_total"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

plt.figure(figsize=(10,6))

sns.barplot(
    data=category_sales,
    x="category",
    y="item_total",
    palette="cubehelix"
)

plt.title("Revenue by Product Category", fontsize=16, fontweight="bold")
plt.xlabel("Category")
plt.ylabel("Revenue")

plt.ticklabel_format(style='plain', axis='y') 

plt.xticks(rotation=45)

plt.tight_layout()

plt.savefig("category_revenue.png", dpi=300)

plt.show()
```

```python
customer_spending = (
    df.groupby("user_id")["item_total"]
    .sum()
)

plt.figure(figsize=(10,6))

sns.histplot(
    customer_spending,
    bins=30,
    kde=True,
    color="steelblue"
)

plt.title("Customer Spending Distribution", fontsize=16, fontweight="bold")
plt.xlabel("Total Spending")
plt.ylabel("Number of Customers")

plt.tight_layout()

plt.savefig("customer_spending_distribution.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="order_status",
    palette="Set2"
)

plt.title("Orders by Status", fontsize=16, fontweight="bold")
plt.xlabel("Order Status")
plt.ylabel("Number of Orders")

plt.tight_layout()

plt.savefig("order_status.png", dpi=300)

plt.show()
```

```python
category_sales = (
    df.groupby("category")["item_total"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,8))

explode = [0.1] + [0]*(len(category_sales)-1)  # first slice explode

plt.pie(
    category_sales,
    labels=category_sales.index,
    autopct='%1.1f%%',
    startangle=140,
    explode=explode,
    shadow=True,
    colors=sns.color_palette("viridis", len(category_sales)),
    wedgeprops={"edgecolor":"white"}
)

plt.title("Revenue Share by Product Category", fontsize=16, fontweight="bold")

plt.tight_layout()

plt.savefig("category_revenue_pie.png", dpi=300)

plt.show()
```
